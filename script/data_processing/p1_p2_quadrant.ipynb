```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
ccri_file = '/content/drive/MyDrive/CCRI/CCRR_repo/data/misc/p1_p2_avg_ccri.csv'
```

```python
ccri_values = pd.read_csv(ccri_file)
```

```python

# Clean iso3
ccri_values['iso3'] = ccri_values['iso3'].astype(str).str.strip()

# Extract P1 and P2
ccri_df = ccri_values[['iso3', 'P1_geometric_avg', 'P2_arithmetic_avg']].dropna()
ccri_df.rename(columns={'iso3': 'ISO3', 'P1_geometric_avg': 'P1', 'P2_arithmetic_avg': 'P2'}, inplace=True)

# 2️⃣ Compute thresholds (median)
ccri_df['P1'] = pd.to_numeric(ccri_df['P1'], errors='coerce')
ccri_df['P2'] = pd.to_numeric(ccri_df['P2'], errors='coerce')

x_thr = ccri_df['P1'].median()
y_thr = ccri_df['P2'].median()

# 3️⃣ Assign quadrant colors
def quadrant_class(x, y):
    if x >= x_thr and y >= y_thr:
        return 'High P1, High P2'    # High P1, High P2
    elif x >= x_thr and y < y_thr:
        return 'High P1, Low P2' # High P1, Low P2
    elif x < x_thr and y >= y_thr:
        return 'Low P1, High P2'  # Low P1, High P2
    else:
        return 'Low P1, Low P2'   # Low P1, Low P2

ccri_df['Quadrant'] = ccri_df.apply(lambda r: quadrant_class(r['P1'], r['P2']), axis=1)
```
