```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
# Specify RP
rp = "100"
```

```python

# URL of the webpage containing links
url = "https://jeodpp.jrc.ec.europa.eu/ftp/jrc-opendata/CEMS-GLOFAS/flood_hazard/RP{}/".format(rp)

# Send a GET request to fetch the webpage
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Create a directory for the downloads
download_dir = "/content/RP{}".format(rp)
os.makedirs(download_dir, exist_ok=True)

# Loop over all the <a> tags to get file links
for link in soup.find_all("a"):
    file_url = link.get("href")
    # Only download files with certain extensions (e.g., .zip, .tif)
    if file_url.endswith(".zip") or file_url.endswith(".tif"):
        full_url = url + file_url
        local_filepath = os.path.join(download_dir, file_url)

        print(f"Downloading: {full_url}")

        # Download the file
        try:
          with requests.get(full_url, stream=True) as r:
              r.raise_for_status()
              with open(local_filepath, 'wb') as f:
                  for chunk in r.iter_content(chunk_size=8192):
                      f.write(chunk)
        except requests.exceptions.HTTPError as e:
            print(f"HTTP Error: {e}")
        except requests.exceptions.RequestException as e:
            print(f"Request Error: {e}")

print("Download complete!")
```
