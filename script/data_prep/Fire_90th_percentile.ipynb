```python
import os
import rasterio
import numpy as np

# Define file path
raster_path = "/content/drive/MyDrive/CCRI/fire/FIRMS_MODIS_Mean_Annual_Count_2001_2023.tif"
output_path = "/content/drive/MyDrive/CCRI/fire/FIRMS_count_90th_percentile.tif"

# Open the raster
with rasterio.open(raster_path) as src:
    data = src.read()  # shape: (bands, rows, cols)
    profile = src.profile

# Mask zeros or negatives as invalid
data = np.where(data > 0, data, np.nan)

# Compute 90th percentile per pixel over time (axis=0)
percentile_90 = np.nanpercentile(data, 90, axis=0)

# Update profile to a single band
profile.update(count=1, dtype='float32', nodata=np.nan)

# Write the output raster
with rasterio.open(output_path, 'w', **profile) as dst:
    dst.write(percentile_90.astype('float32'), 1)
```
