```python
import xarray as xr
import glob
import rioxarray
import os

# Load all .nc files
files = sorted(glob.glob(f"ecmwf-data_12/{layer}12*.nc"))

# Open all files
ds = xr.open_mfdataset(files, combine='by_coords')

# Compute 10th percentile across the 'time' dimension
period_percentile = ds[f'{layer}12'].quantile(0.1, dim='time')

# Set spatial dimensions
period_percentile = period_percentile.rio.set_spatial_dims(x_dim="lon", y_dim="lat")

# Assign CRS (Coordinate Reference System)
period_percentile = period_percentile.rio.write_crs("EPSG:4326")

# Export to GeoTIFF
period_percentile.rio.to_raster(f"{layer}12_period_10thpercentile_2014_2024.tif")
```
